```python
import numpy as np
eig_matrix = standard
eig_matrix = np.nan_to_num(eig_matrix)
#特征归一化后再进行降维
import numpy.linalg as LA
eig_matrix_cov = np.cov(eig_matrix)#返回协方差矩阵
w,v = LA.eig(eig_matrix_cov)#计算特征值和特征向量
eig_matrix_pca = np.matmul(eig_matrix.T, v[:,:64]) #n->64
eig_matrix_pca = eig_matrix_pca.real #只保留实数部分
eig_matrix_pca
```

```python
def sparse_corr_analysis(adj_matrix, C=1.0, A=1.0, iterations=50, beta=0.8, K=5, threshold=1e-5, dim=3):
    nodes_num = adj_matrix.shape[0]
    adj_matrix = adj_matrix.tolil()
    adj_matrix = adj_matrix.astype('float32')
    #init_pos = np.asarray(np.random.randn(nodes_num, dim), dtype=adj_matrix.dtype)
    init_pos = np.nan_to_num(df_fv.to_numpy(dtype=adj_matrix.dtype))#降维后的特征向量
    #print('Init pos', init_pos)

    l = C * np.sqrt(A / nodes_num)
    displacement = np.zeros((dim, nodes_num))
    pos = init_pos.copy()
    
    for iteration in range(iterations):
        displacement *= 0
        # learning rate
        lr = beta**(iteration / K)
        for i in range(adj_matrix.shape[0]):
            delta = (pos[i] - pos).T
            distance = np.sqrt((delta ** 2).sum(axis=0))
            distance = np.where(distance < 0.01, 0.01, distance)
            Ai = np.asarray(adj_matrix.getrowview(i).toarray())
            # print('Ai', Ai) # adjacency matrix of node i
            displacement[:, i] += \
                (delta * (l**2 / distance - Ai * distance**2 / l)).sum(axis=1)
        
        # update positions
#         length = np.sqrt((displacement ** 2).sum(axis=0))
        length = np.linalg.norm(displacement, axis=0)
        length = np.where(length < 0.01, 0.1, length)
        delta_pos = (displacement * lr / length).T
        pos += delta_pos
        
        # error
        err = np.linalg.norm(delta_pos) / nodes_num
        if err < threshold:
            break
   
    return (init_pos, pos, err)
```
