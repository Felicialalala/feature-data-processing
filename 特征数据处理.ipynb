```python
#计算样本间距
import itertools
data_numpy = df.to_numpy()
distances = []
```

```python
for i in range(0, len(data_numpy[0])):
    distance = 0
    for x,y in itertools.combinations(data_numpy[:, i],2):
        distance += abs(x-y)
    mean_distance = (2.0/(len(data_numpy[:, i])*(len(data_numpy[:, i])-1)))*distance
    distances.append(mean_distance)
distances 
```

```python
#对数据进行归一化(最大最小值)
'''
Parameters: 
    dataSet - 特征矩阵
Returns:
    normDataSet - 归一化后的特征矩阵
    ranges - 数据范围
    minVals - 数据最小值
'''
def autoNorm(dataSet):
    minVals = dataSet.min(0)
    maxVals = dataSet.max(0)
    ranges = maxVals - minVals
    row = dataSet.shape[0]
    #原始值减去最小值除以最大值和最小值的差，得到归一化数据
    normDataSet = (dataSet - np.tile(minVals,(row,1))) / np.tile(ranges,(row,1)) 
    return normDataSet, ranges, minVals
```
